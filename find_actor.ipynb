```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('[DS]TermProject_data.csv', encoding='cp949')
df.head()
```

```python
import requests
from bs4 import BeautifulSoup as bs

df_actor = pd.DataFrame({
        '영화명': ["initial_movie"],
        '감독': ["initial_director"],
        '배우': ["initial_actor"]
})


for i in range(0,len(df)):
        try:
                page = requests.get(f"http://www.cine21.com/search/movie/?q={df.at[i,'영화명']}")
                soup = bs(page.text, "html.parser")
                print(soup.find(class_='mov_list').find(class_='name').find("a").text)

                for movieLi in soup.find(class_='mov_list').find_all("li"):
                        sub_infos = movieLi.find_all(class_='sub_info')
                        sub_info = sub_infos[1]
                        
                        try:
                                
                                if(df.at[i,'감독']==sub_info.find("a").text):
                                        for peopleLi in sub_info.find_all("a"):
                                                if(peopleLi.text!=df.at[i,'감독']):
                                                        print("ok")
                                                        df_actor1=pd.DataFrame({
                                                        '영화명': [df.at[i,'영화명']], 
                                                        '감독': [df.at[i,'감독']], 
                                                        '배우': [peopleLi.text]
                                                        })
                                                        
                                                        df_actor = pd.concat([df_actor, df_actor1], ignore_index=True)
        
                                        break
                        except:
                                continue
                        
                

        except:
                continue 
df_actor.drop(labels=0,axis=0,inplace=True)
df_actor.to_csv("C:/Users/leeminwoo/Desktop/데과팀플/df_actor.csv")
```
